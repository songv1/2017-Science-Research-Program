```python
#BHAR vs. probability for each one, machine
from astropy.io import fits, ascii
import matplotlib.pyplot as plt
from astropy.table import Table, join
import numpy as np
%config InlineBackend.rc = {}
import matplotlib
matplotlib.rc_file("anaconda/lib/python2.7/site-packages/matplotlib/mpl-data/matplotlibrc")
%matplotlib inline

#Setting up the figure
fig = plt.figure()
fig.set_size_inches(14,7)
fig.suptitle('BHAR vs. Object Type Probability', fontsize=20)

t = ascii.read('gds_allz.txt')
ax = []

#Setting up the grid
for i in range(2):
    ax.append(fig.add_subplot(1, 2, i+1))
    if i == 0: 
        ax[i].set_title('Disk', fontsize=20)
        ax[i].set_ylabel('Disk Probability', fontsize=20)
        ax[i].scatter(t['BHAR'], t['visualHCPG_f_disk'], s=40, facecolor=(0,0,1))
    else:
        ax[i].set_title('Point Source', fontsize=15)
        ax[i].set_ylabel('Point Source Probability', fontsize=20)
        ax[i].scatter(t['BHAR'], t['visualHCPG_f_ps'], s=40, facecolor=(1,0,0))
    
    ax[i].set_xlabel('BHAR in log $M_{\odot}$ ${year^{-1}}$', fontsize=20)
    ax[i].set_xlim(-2.5, 0)
    ax[i].set_ylim(0, 1)

    #Tick mark label size
    ax[i].tick_params(axis='both', which='major', labelsize=15)

    ax[i].axhline(y=0.5,xmin=-10,xmax=10,c="black",linewidth=3,zorder=0)
    
#Tune spacing between plots. 
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.3, hspace=None)
    
fig.savefig('bhar_machine_line.png', dpi=300, bbox_inches='tight')
```
